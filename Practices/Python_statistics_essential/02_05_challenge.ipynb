```python
import numpy as np
import pandas as pd

import matplotlib
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
data = pd.read_csv('tb.csv')
data.head(20)
```

```python
data.columns
```

```python
melted = pd.melt(data,id_vars=['country','year'],value_vars=['m04', 'm514', 'm014', 'm1524', 'm2534', 'm3544',
       'm4554', 'm5564', 'm65', 'mu', 'f04', 'f514', 'f014', 'f1524', 'f2534',
       'f3544', 'f4554', 'f5564', 'f65', 'fu'],var_name='sexage',value_name='cases').dropna()
```

```python
melted['sex'] = melted.sexage.str[0]
melted['age'] = melted.sexage.str[1:]
```

```python
melted.age = melted.age.map({'04':'0-4',
                                '514':'5-14',
                                '014':'0-14',
                                '1524':'15-24',
                                '2534':'25-34',
                                '3544':'35-44',
                                '4554':'45-54',
                                '5564':'55-64',
                                'u':'65up',
                                })
```

```python
melted = melted.drop('sexage',axis=1)
melted.cases = melted.cases.apply(int)
```

```python
melted = melted[['country','year','sex','age','cases']].sort_values(['country','year','sex','age','cases']).dropna()
```
